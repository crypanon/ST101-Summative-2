```python
from image_util import load, show, repetition
import numpy as np
```

### Testing show() function

```python
def test_show():
    # Test showing a valid image
    test_img = np.random.randint(0, 256, size=(300, 300, 3), dtype=np.uint8)
    try:
        show(test_img)
    except TypeError:
        pass

    # Test showing an invalid image (empty array)
    try:
        show(np.zeros((0, 0, 3), dtype=np.uint8))
    except ValueError:
        pass
```

### Testing repetition() function

```python
def test_repetition():
    # Test repetition with valid inputs
    test_img = np.random.randint(0, 256, size=(50, 50, 3), dtype=np.uint8)
    show(test_img)
    repeated_img = repetition(test_img, 2, 3)
    assert repeated_img.shape == (100, 150, 3)
    

    # Test repetition with invalid inputs (non-integer rows and cols)
    try:
        repetition(test_img, 2.5, 3)
    except ValueError:
        pass

    try:
        repetition(test_img, 2, 3.5)
    except ValueError:
        pass

    # Test repetition with invalid input (empty array)
    try:
        repetition(np.zeros((0, 0, 3), dtype=np.uint8), 2, 3)
    except ValueError:
        pass
```

```python
test_show()
test_repetition()
```
